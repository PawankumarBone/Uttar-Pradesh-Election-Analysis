# 7. In how many constituencies, the difference in Total Votes between the Bharatiya Janata Party and the Samajwadi Party was less than 5000.

```python
jk
```

```python
jk2=jk[(jk['Party']=='Bharatiya Janata Party')|(jk['Party']=='Samajwadi Party')][['ac_no','Total Votes']].reset_index()
jk2.drop('index',axis=1,inplace=True)
jk2
```

```python
jk3={}
for i in range(0,len(jk2)-1,2):
    jk3[jk2['ac_no'][i]]=abs(jk2['Total Votes'][i]-jk2['Total Votes'][i+1])
jk3
```

```python
jk4=pd.DataFrame(list(jk3.items()),columns = ['ac_no','Vote_diff'])
jk4
```

```python
jk4[jk4['Vote_diff']<5000]['ac_no'].count()
```

```python
print(jk4[jk4['Vote_diff']<5000]['ac_no'].count(),"Number of constituencies the difference in Total Votes between the Bharatiya Janata Party and the Samajwadi Party was less than 5000")
```
